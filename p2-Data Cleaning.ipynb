```python
import pandas as pd
import numpy as np
import re
```

```python
df = pd.read_excel("hnrealestate.xlsx")
```

```python
df.drop_duplicates(subset=['Road_width', 'Area','Size','Direction','Price','Location'],keep='first',inplace=True)
df
```

```python
# Clean the Road_width column 

road_width = []
for e in df["Road_width"]:
    a = re.findall(r"[-+]?(?:\d*\.\d+|\d+)",str(e.replace('m', '').replace('-', 'NaN').replace(',','.')))
    if len(a) == 1:
        road_width.append(float(a[0]))
    else:
        road_width.append(np.NaN)
df["Road_width"] = road_width
```

```python
df
```

```python
# Use "Size" column to create "width" and "length" from "Size" for each unit
size = []
length = []
width = []
for e in df["Size"]:   
    e = e.replace('m','').replace('---','0').replace(',','.')
    try: 

        size.append(e)
        a = float(e.split("x")[0])
        b = float(e.split("x")[1])
        if a >= b:
            length.append(a)
            width.append(b)
        else:
            length.append(b)
            width.append(a)
    except:
        length.append(0)
        width.append(0)
length = [np.NAN if e == 0 else e for e in length]
width = [np.NAN if e == 0 else e for e in width]
df["Length"] = length
df["Width"] = width
df.drop(columns = ['Size'],inplace = True)
```

```python
df
```

```python
# Clean "Area" column
area = []
x = 0
for e in df["Area"]:
    try:
        a = float(e.replace(' m2',''))
    except:
        try:
            a = float(df["Length"].values[x] * df["Width"].values[x])
        except:
            a = np.NaN
    area.append(a)
    x += 1
len(area)   
df['Area'] = area 
```

```python
# Clean "Price" column
list(df.Price.unique())
```

```python
# Clean "Price" variable
prices= []
count=0
for e in df.Price:
    new_e = e.replace(',','.')
    
    if "tỷ" in new_e:
        price = float(re.findall(r"[-+]?(?:\d*\.\d+|\d+)",new_e)[0])
        prices.append(price)
    elif "triệu /" in new_e:
        if "m2" in new_e:
            price1 = float(df.iloc[count]["Area"])*float(re.findall(r"[-+]?(?:\d*\.\d+|\d+)",new_e)[0])/1000
            prices.append(price1)
        else: 
            price2 = float(re.findall(r"[-+]?(?:\d*\.\d+|\d+)",new_e)[0])/1000 
            prices.append(price2)
    else:
        prices.append(np.NaN)
    count+=1
prices
```
