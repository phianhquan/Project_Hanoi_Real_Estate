```python
import pandas as pd
import numpy as np
from math import sin, cos, sqrt, atan2, radians, asin
import numpy as np
```

```python
df = pd.read_excel("hn_real_estate_22.xlsx")
df
```

```python
df_loca = pd.read_excel("Landmark.xlsx")
```

```python
def haversine(lat1, lon1, lat2, lon2):
    """
    Calculate the great circle distance in kilometers between two points 
    on the earth (specified in decimal degrees)
    """
    # convert decimal degrees to radians 
    lon1, lat1, lon2, lat2 = map(radians, [lon1, lat1, lon2, lat2])

    # haversine formula 
    dlon = lon2 - lon1 
    dlat = lat2 - lat1 
    a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
    c = 2 * asin(sqrt(a)) 
    r = 6371
    return c * r
```

```python
###Create a column having the distance to the center of the city

df["Distance from center(km)"] = df.apply(lambda row: haversine(row['Latitude'], row['Longitude'], 21.027671, 105.834168), axis=1)
df
```

```python
def distance_column():   
    """
    Create a column having the distance to the nearest major road.
    """
    # Convert lat and lon to float
    df["Latitude"].apply(lambda x: float(x))
    df["Longitude"].apply(lambda x: float(x))

    site = df[["Latitude","Longitude"]]

    # Creating the list of distance values for the new column
    dist_col = []    
    for i1 in range(len(site)):
        dist_list = []
        lat1 = site.iloc[i1]["Latitude"]
        lon1 = site.iloc[i1]["Longitude"]
        for i2 in range(len(df_loca)):
            lat2 = df_loca.iloc[i2]["latitude"]
            lon2 = df_loca.iloc[i2]["longitude"]
            dist = haversine(lat1,lon1,lat2,lon2)
            dist_list.append(dist)
        dist_col.append(min(dist_list))
    df["distance"] = dist_col

distance_column()
```

```python
### Remove outlier value in the distance column
dis=[]
for i in range(len(df)):
    if df.iloc[i]["Distance from center(km)"] <= df.iloc[i]["distance"]:
        dis.append(df.iloc[i]["Distance from center(km)"])
    else:
        dis.append(df.iloc[i]["distance"])
df["Distance from nearest road (km)" ] = dis
```
